This notebook explores multi-channel digital advertising performance data. The goal is to:
- Inspect the dataset for quality and structure
- Compute key digital marketing metrics (CTR, ROAS, etc.)
- Prepare clean data for further analysis and dashboarding

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Show all columns when previewing
pd.set_option('display.max_columns', None)

# Load the dataset
df = pd.read_csv('../data/media_campaign_data.csv')

# Preview first few rows
df.head()
```

## üìê Basic Dataset Info

Let‚Äôs check:
- Number of rows and columns
- Column data types
- Non-null counts

```python
# Shape of the dataset
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# Structure and datatypes
df.info()
```

## üìä Summary Statistics

Let‚Äôs view the statistical summary of numerical columns to check ranges and distributions.
This helps identify extreme values, zeros, or data entry errors.


```python
df.describe()
```

## üßπ Missing Values Check

Let‚Äôs verify that there are no missing values in critical columns like Spend, Impressions, Conversions, etc.

```python
df.isnull().sum()
```

## üß† Campaign and Channel Breakdown

Let‚Äôs see:
- How many platforms are in the dataset?
- Which campaigns exist and how often they appear?

```python
# Unique values per channel
print("Channels:")
print(df['Channel'].value_counts(), '\n')

# Number of unique campaigns
print(f"Number of unique campaigns: {df['Campaign Name'].nunique()}")
```

## üìà Derived Metrics (CTR, CPC, CPA, ROAS)

Let‚Äôs calculate key performance indicators for campaign performance:
- **CTR (Click-Through Rate)**
- **CPC (Cost Per Click)**
- **CPA (Cost Per Acquisition)**
- **ROAS (Return on Ad Spend)**

We'll store these as new columns in the dataset.


```python
# Avoid division errors by replacing 0 with NaN
df.replace(0, np.nan, inplace=True)

# Add calculated KPI columns
df['CTR'] = df['Clicks'] / df['Impressions']
df['CPC'] = df['Spend'] / df['Clicks']
df['CPA'] = df['Spend'] / df['Conversions']
df['ROAS'] = df['Revenue'] / df['Spend']

# Preview new columns
df[['CTR', 'CPC', 'CPA', 'ROAS']].describe()
```

## üßΩ Handle Invalid or Infinite Values

We‚Äôll clean up any invalid values in the calculated metrics caused by:
- Division by 0 (e.g. no clicks or no conversions)
- Infinite or undefined results

This is a key step before moving into visualization or dashboarding.

```python
# Replace infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows with NaN values in key metrics
df.dropna(subset=['CTR', 'CPC', 'CPA', 'ROAS'], inplace=True)

# Check final dataset shape
print(f"Cleaned rows remaining: {df.shape[0]}")
```

## üíæ Save Cleaned Dataset

Now that we've calculated and cleaned our key KPIs, we‚Äôll save the cleaned version of the dataset into the `data/clean/` folder for future analysis and dashboarding.


```python
# Save to /data/clean/
df.to_csv('../data/clean/media_campaign_data_clean.csv', index=False)
print("‚úÖ Cleaned dataset saved successfully.")
```

## üìä Univariate Visualizations

Let‚Äôs examine the distribution of key numeric metrics:
- Spend
- Impressions
- Clicks
- Conversions
- Revenue

This helps us spot any skew, anomalies, or outliers in the data before aggregating or modeling.

```python
metrics = ['Spend', 'Impressions', 'Clicks', 'Conversions', 'Revenue']
plt.figure(figsize=(16, 10))

for i, metric in enumerate(metrics, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[metric], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {metric}')
    plt.xlabel(metric)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## üíπ Derived Metric Distributions (CTR, CPC, CPA, ROAS)

Next, we explore how our calculated KPIs are distributed to identify what's "typical" campaign performance across platforms.


```python
kpis = ['CTR', 'CPC', 'CPA', 'ROAS']
plt.figure(figsize=(16, 8))

for i, kpi in enumerate(kpis, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[kpi], kde=True, bins=30, color='coral')
    plt.title(f'Distribution of {kpi}')
    plt.xlabel(kpi)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## üìà ROAS by Channel

Let‚Äôs analyze which platforms delivered the best Return on Ad Spend (ROAS). We'll use box plots to compare the distribution across channels.


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Channel', y='ROAS', data=df, hue='Channel', palette='Set2', legend=False)
plt.title('ROAS Distribution by Channel')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig('../images/roas_by_channel.png', dpi=300)
plt.show()
```

## üß≠ CTR vs. ROAS

This scatter plot shows if campaigns that get more clicks (higher CTR) also tend to convert better (ROAS). Helps distinguish between traffic quality vs. volume.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='CTR', y='ROAS', hue='Channel', data=df, palette='tab10')
plt.title('CTR vs ROAS by Channel')
plt.xlabel('Click-Through Rate (CTR)')
plt.ylabel('Return on Ad Spend (ROAS)')
plt.grid(True)
plt.tight_layout()
plt.savefig('../images/ctr_vs_roas.png', dpi=300)
plt.show()
```

# üìå Business Questions & Strategic Insights

Now that we‚Äôve cleaned and explored the media campaign data, we‚Äôll answer several business-critical questions:

1. Which channels perform best in terms of Return on Ad Spend (ROAS)?
2. Which campaigns are most efficient at acquiring customers (lowest CPA)?
3. Are high CTR campaigns also high ROAS? (i.e. does traffic quality = business value?)
4. How consistent is performance across channels?


## üí∏ Q1: Which channels had the highest average ROAS?

Let‚Äôs group the data by channel and compare their average ROAS.

```python
channel_roas = df.groupby('Channel')['ROAS'].agg(['mean', 'median', 'std', 'count']).sort_values(by='mean', ascending=False)
channel_roas.style.background_gradient(cmap='Greens')
```

## üéØ Q2: Which campaigns are most cost-effective (lowest CPA)?

This helps prioritize which campaigns bring in the most conversions at the lowest price.

```python
top_cpa = df.groupby('Campaign Name')['CPA'].agg(['mean', 'count']).sort_values('mean')
top_cpa.head(10).style.background_gradient(cmap='Blues')
```

## ü§î Q3: Is high CTR associated with high ROAS?

We‚Äôll check the correlation between CTR and ROAS.

```python
correlation = df[['CTR', 'ROAS']].corr().iloc[0, 1]
print(f"Correlation between CTR and ROAS: {correlation:.2f}")
```

## üìâ Q4: How consistent is performance across platforms?

Standard deviation of ROAS per channel helps identify **volatile** vs **reliable** platforms.


```python
channel_roas_std = channel_roas[['mean', 'std']].sort_values('std')
channel_roas_std.style.background_gradient(cmap='YlOrRd')
```

# ‚úÖ Strategic Recommendations

Based on our analysis, here are key insights for media optimization:

- **High ROAS Channels**: Meta Ads and TikTok delivered the highest average ROAS, indicating strong return on investment. Prioritize spend on these platforms.
- **Cost-Efficient Campaigns**: Campaigns such as Lookalike Expansion and Creator Collab achieved the lowest average CPA, making them highly efficient at acquiring users.
- **CTR ‚â† ROI**: Click-through rate and ROAS showed only a moderate correlation. High engagement does not always translate to profitability ‚Äî evaluate traffic quality alongside conversion value.
- **Stable Performers**: YouTube and LinkedIn had the most consistent ROAS (lowest variability), making them strong candidates for predictable, steady performance.
